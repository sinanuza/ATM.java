```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier ,export_graphviz
import graphviz
from sklearn.model_selection import train_test_split

```

```python
df=pd.read_csv('heart.csv')
```

```python
df.head(3)
```

```python
y=df['output']
x=df.drop('output',axis=1)
```

```python
tree=DecisionTreeClassifier()
model=tree.fit(x,y)
```

```python
model.score(x,y)
```

```python
df.info()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=16,test_size=0.70)
```

```python
tree=DecisionTreeClassifier()
model=tree.fit(x_train,y_train)
model.score(x_test,y_test)
```
